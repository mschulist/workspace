```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
winners = pd.read_csv("results.csv")
winners.columns
```

```python
winners[(winners["p_no_winner"] == 0) & (winners["n_voters"] != 1)].sort_values("n_alternatives", ascending=False)
```

```python
fig = plt.figure(figsize=(5, 6))

# syntax for 3-D projection
ax = plt.axes(projection="3d")

ax.scatter3D(winners["n_voters"], winners["n_alternatives"], winners["p_no_winner"], "blue")

ax.set_xlabel("Number of voters")
ax.set_ylabel("Number of alternatives")
ax.set_zlabel("Probability of no winner")
ax.set_box_aspect(None, zoom=0.9)

plt.title(
    "Number of voters, number of alternatives,\nand probability of no winner in Concordet voting system"
)

plt.show()
```
