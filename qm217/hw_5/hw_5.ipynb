```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.integrate as integrate
```

```python
alpha = 3

integrate.quad(lambda x: np.sqrt(alpha / 2) ** 2 * np.exp(- alpha * np.abs(x)), a=-1000, b=1000)
```

```python
def psi_re(x):
    alpha = 1
    N = np.sqrt(alpha / 2)
    return N * np.exp(-alpha * np.abs(x) / 2) * np.cos(x)


psi_x = np.linspace(-10, 10, 1000)
psi_y = psi_re(psi_x)

with open("psi_re.csv", "w") as f:
    for x, y in zip(psi_x, psi_y):
        f.write(f"{x},{y}\n")

plt.plot(psi_x, psi_y)
plt.title("Real Part of Wave Function")
plt.show()
```

```python
def psi_sq(x):
    alpha = 1
    N = np.sqrt(alpha / 2)
    return N**2 * np.exp(-alpha * np.abs(x))


psi_sq_y = psi_sq(psi_x)

with open("psi_sq.csv", "w") as f:
    for x, y in zip(psi_x, psi_sq_y):
        f.write(f"{x},{y}\n")

plt.plot(psi_x, psi_sq_y)
plt.title("Wave Function Squared")
plt.show()
```

```python
integrate.quad(psi_sq, a=-1000, b=1000)
```

```python
def U(x):
    return (1 / x**2) - (2 / x) + 1


x_star = np.linspace(0.35, 8, 1000)
U_star = U(x_star)

with open("prob_4_plot.csv", "w") as f:
    for i in range(len(x_star)):
        f.write(f"{x_star[i]},{U_star[i]}\n")

plt.plot(x_star, U_star)
plt.title("Potential Energy")
plt.show()
```

```python
def g(x):
    k = 1
    k0 = 2
    alpha = 1
    N = np.sqrt(alpha / 2)
    return N / (np.pi * 2) * np.exp(-alpha * np.abs(x) / 2) * (np.cos(k0 * x) + np.sin(k * x))

integrate.quad(g, a=-1000, b=1000)
```
