```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    out = x * np.tan(x)
    out[np.abs(out) > 15] = np.nan
    out[out < 0] = np.nan
    return out


def z(theta):
    x, y = 10 * np.cos(theta), 10 * np.sin(theta)
    x[x < 0] = np.nan
    y[y < 0] = np.nan
    return x, y


def g(x):
    out = -x * 1 / np.tan(x)
    out[np.abs(out) > 15] = np.nan
    out[out < 0] = np.nan
    return out


x_vals = np.linspace(0, 4 * np.pi, 1000)
y_vals_f = f(x_vals)
y_vals_g = g(x_vals)

theta_vals = np.linspace(0, 0.5 * np.pi, 1000)
x_vals_z, y_vals_z = z(theta_vals)

with open("even_data_eta.csv", "w") as file:
    for x, y in zip(x_vals, y_vals_f):
        if not np.isnan(y) and not np.isnan(x):
            file.write(f"{x},{y}\n")

with open("data_circle.csv", "w") as file:
    for x, y in zip(x_vals_z, y_vals_z):
        if not np.isnan(y) and not np.isnan(x):
            file.write(f"{x},{y}\n")

with open("odd_data_eta.csv", "w") as file:
    for x, y in zip(x_vals, y_vals_g):
        if not np.isnan(y) and not np.isnan(x):
            file.write(f"{x},{y}\n")

plt.plot(x_vals, y_vals_f)
plt.plot(x_vals_z, y_vals_z)
plt.plot(x_vals, g(x_vals))
plt.show()
```
