```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
```

```python
def psi(x):
    a = 1
    k = 2 * np.pi - 0.1
    K = 2.3

    if x < -a:
        return np.exp(K * x)
    elif -a <= x <= a:
        return np.sin(k * x)
    else:
        return -np.exp(-K * x)


x_range = np.linspace(-2, 2, 1000)
y_outs = np.array([psi(x) for x in x_range])
```

```python
with open("prob_2_psi.csv", "w") as f:
    for x, y in zip(x_range, y_outs):
        f.write(f"{x},{y}\n")

plt.plot(x_range, y_outs)
plt.show()
```

```python
def psi_sq(x):
    return psi(x) ** 2


y_sq_outs = np.array([psi_sq(x) for x in x_range])

with open("prob_2_psi_sq.csv", "w") as f:
    for x, y in zip(x_range, y_sq_outs):
        f.write(f"{x},{y}\n")

plt.plot(x_range, y_sq_outs)
plt.show()
```
