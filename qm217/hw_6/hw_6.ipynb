```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def psi_sq(x):
    L = 5
    return 2 / L * np.sin(5 * np.pi * x / L) ** 2

x_vals = np.linspace(0, 5, 1000)
y_vals = psi_sq(x_vals)

with open("prob_5.csv", "w") as f:
    for x, y in zip(x_vals, y_vals):
        f.write(f"{x},{y}\n")

plt.plot(x_vals, y_vals)
plt.show()
```
